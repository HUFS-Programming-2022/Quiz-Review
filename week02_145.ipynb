```python
#fdist로 빈도수 확인하기
fdist = FreqDist(text5)
fdist
```

```python
#리스트 컴프리헨션에서 fdist로 빈도수 4인 단어들만 걸러낸다
print([word for word in text5 if fdist[word]==4])
```

```python
#중복 제거하기
print(set(([word for word in text5 if fdist[word]==4])))
```

```python
#오름차순으로 정렬하기
print(sorted(set(([word for word in text5 if fdist[word]==4]))))
```
