## 1. CFD를 이용하여 nltk.book의 텍스트별 빈도수를 만들고 wh-words의 빈도수를 출력하시오. 
- wh_words = what, why, when, where, which

```python
texts = [text1, text2, text3, text4, text5, text6, text7, text8, text9]
wh_words = ['what', 'why', 'when', 'where', 'which']

# 텍스트 리스트의 각 텍스트를 돈다
# wh-words리스트의 각 wh-word를 돈다
# 텍스트의 단어 중 wh-word인 것을 찾는다
cfd = nltk.ConditionalFreqDist(
    (text, wh)
    for text in texts
    for wh in wh_words
    
    for word in text
        if word.lower() == wh
)
```

```python
# 각 텍스트를 돌며 cfd를 이용해 wh-word를 찾아 출력한다
# end=' '와 print()를 이용해 출력 결과의 틀을 맞춘다
for text in texts:
    print(f'{text}', end=' ')
    for wh in wh_words:
        print(f'{wh}:{cfd[text][wh]}번', end=' ')
    print()
```

## 2. gutenberg에서...
- 파일 중 모음(a, e, i, o, u)으로 시작하는 파일을 골라
- 단어를 가져오되 숫자, 문장부호를 모두 제거하고
- 모든 단어를 하나의 스트링으로 출력하시오.

```python
# 리스트 컴프리헨션을 이용해 모음으로 시작하는 파일을 찾는다
files = [file for file in gutenberg.fileids() if file.startswith(('a', 'e', 'i', 'o', 'u'))]
```

```python
# 리스트 컴프리헨션을 이용해 각 파일의 단어 중 알파벳인 것만 찾는다
words = [[word for word in gutenberg.words(file) if word.isalpha()] for file in files]
words[:1000]
```

```python
# join을 이용해 텍스트 안의 단어들을 합쳐준 뒤 각 텍스트를 합쳐 하나의 스트링으로 출력한다
result = ', '.join([', '.join(text) for text in words])
result[:600]
```

## 3. 브라운에서 각 장르별 hapax의 개수를 cfd를 이용하여 구하시오.

```python
# 카테고리의 각 장르를 돈다
# 장르의 각 단어를 돈다
cfd = nltk.ConditionalFreqDist(
    (genre, word.lower())
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)

# 각 장르를 돌며 cfd를 이용해 장르의 hapax의 개수를 구한다
for genre in brown.categories():
    print(f'{genre}의 hapax => {len(cfd[genre].hapaxes())}개')
```
