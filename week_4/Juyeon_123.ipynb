```python
"""
1. CFD를 이용하여 nltk.book의 텍스트별 빈도수를 만들고 wh-words의 빈도수를 출력하시오. wh_words = what, why, when, where, which
"""
#1. nltk.book의 텍스트별 wh-words의 빈도수를 구하고 싶을 때
def carc_cfd_wh_words():
    import nltk
    from nltk import book
    nltk.download('book', quiet = True)#
    
    texts = [getattr(nltk.book, f'text{i}') for i in range(1, 10)]
    wh_words = ['what', 'why', 'when', 'where', 'which']

    cfd = nltk.ConditionalFreqDist(
        (text, word.lower())
        for text in texts
        for whw in wh_words
        for word in text
            if word.lower() == whw
    )    
    
    for text in texts:
        print(f'{text}에서')
        for word in wh_words:
            print(f'{word}의 빈도수는 {cfd[text][word]}번')
    

carc_cfd_wh_words()
```

```python
#2. nltk.book의 특정 텍스트 wh-words의 빈도수를 구하고 싶을 때
def carc_cfd_wh_words_particular(text):
    import nltk
    from nltk import book
    nltk.download('book', quiet = True)#
        
    wh_words = ['what', 'why', 'when', 'where', 'which']

    cfd = nltk.ConditionalFreqDist(
        (text, word.lower())
        for whw in wh_words
        for word in text
            if word.lower() == whw
    )    
    
    print(f'{text}에서')
    for word in wh_words:
        print(f'{word}의 빈도수는 {cfd[text][word]}번')
    

carc_cfd_wh_words_particular('Text: Moby Dick by Herman Melville 1851')

#이 코드에서는 빈도수가 제대로 세어지지 않는데, 스스로 이유를 알 수가 없어 그대로 올립니다! 
```

```python
"""
2. gutenberg에서...
    -파일 중 모음(a, e, i, o, u)으로 시작하는 파일을 골라
    -단어를 가져오되 숫자, 문장부호를 모두 제거하고
    -모든 단어를 하나의 스트링으로 출력하시오.
"""
#1. gutenberg의 모든 파일에서 구하고 싶을 때
def refine_gutenberg():
    import nltk 
    from nltk.corpus import gutenberg
    nltk.download('gutenberg', quiet = True)
    
    vowel = ['a', 'e', 'i', 'o', 'u']
    
    for files in gutenberg.fileids():
        if files[0] in vowel:
            word =  [word.strip('123456789').strip(',.:;"~!?-*').strip('\'') for word in gutenberg.words(files)]
            word_string = ' '.join(word)
            
    print(word_string)
            
            
refine_gutenberg()
```

```python
#2. gutenberg의 특정 파일에서 구하고 싶을 때
def refine_gutenberg_particular(files):
    import nltk 
    from nltk.corpus import gutenberg
    nltk.download('gutenberg', quiet = True)
    
    vowel = ['a', 'e', 'i', 'o', 'u']
    
   
    if files[0] in vowel:
        word =  [word.strip('123456789').strip(',.:;"~!?-*()').strip('\'') for word in gutenberg.words(files)]
        word_string = ' '.join(word)
            
    print(word_string)
            
            
refine_gutenberg_particular('austen-emma.txt')
```

```python
"""
3. 브라운에서 각 장르별 hapax의 개수를 cfd를 이용하여 구하시오.
"""
#1. 모든 장르의 hapax를 구하고 싶을 때
def carc_brown_hapaxes():
    import nltk

    nltk.download('brown')
    brown_corpus = nltk.corpus.brown

    cfd = nltk.ConditionalFreqDist(
        (genre, word)
        for genre in brown_corpus.categories()
        for word in brown_corpus.words(categories = genre)
        )
    
    for genre in brown_corpus.categories():
        print(f'{genre} hapaxes : {len(cfd[genre].hapaxes())}개')

carc_brown_hapaxes()

#2. 특정 장르의 hapax를 구하고 싶을 때

print('고를 수 있는 장르명은 다음과 같습니다!')
for genre in brown_corpus.categories():
    print(f'{genre}')   
        
def carc_brown_hapaxes_particular(genre = input('장르명을 입력해 주세요!')):
    import nltk

    nltk.download('brown')
    brown_corpus = nltk.corpus.brown
        
    cfd = nltk.ConditionalFreqDist(
        (genre, word)
        for genre in brown_corpus.categories()
        for word in brown_corpus.words(categories = genre)
        )
    
    print(f'{genre} hapaxes : {len(cfd[genre].hapaxes())}개')
    
carc_brown_hapaxes_particular()
```
