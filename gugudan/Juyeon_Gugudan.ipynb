## 구구단 계산기
- 사용자에게 메뉴를 입력받고 그에 해당하는 결과물을 출력한다.

- 프로그램은 사용자가 ‘q’을 입력했을 때에만 중지된다.

- 조건에 맞지 않는 입력값이 들어왔을 때 재확인하기
- 어떤 상황에서도 프로그램이 에러 없이, 오류 없이 작동하도록 신경쓰기
- 함수로 코드 정리하기

```python
"""
메뉴:

구구단 출력기
----------------------------------------------------------------------
           1) n단 출력    2) n단까지 출력   3) n ~ m단 출력   q) 나가기
----------------------------------------------------------------------
메뉴를 선택하세요 > 
각 메뉴에 맞는 출력물을 프린트합니다.

조건1: 0 < n < 10
조건2: n ≤ m
조건3: 모든 상황에서 q을 눌러 종료할 수 있어야 한다.
조건4: 종료 시 출력할 문구 “이용해주셔서 감사합니다.”
예외 상황에 대처 잘하기
"""
```

```python
# 1. 기본 함수 만들기

# 한 단 구구단 함수
def calc_single_times_table(num):
    for num in range(num, num+1):
        print(f'== {num}단 ==')
        for i in range(1, 10):
            print(f'{num} x {i} = {num*i}')
            
# n단까지 출력 구구단 함수
def calc_ton_times_table(num):
    for num in range(1, num+1) :
        print(f'== {num}단 ==')
        for i in range(1, 10) :
            print(f'{num} x {i} = {num*i}')
            
# n ~ m단 출력 구구단 함수
def calc_1to2_times_table(num1, num2):
    for num1 in range(num1, num2+1) :
        print(f'== {num1}단 ==')
        for i in range(1, 10) :
            print(f'{num1} x {i} = {num1*i}')
            
# n ~ m단 출력 구구단 함수_숫자 변경 ver
def calc_2to1_times_table(num2, num1):
    for num2 in range(num2, num1+1) :
        print(f'== {num2}단 ==')
        for i in range(1, 10) :
            print(f'{num2} x {i} = {num2*i}')   
        
# 프로그램 종료 시 출력될 함수
def end():
    print("이용해 주셔서 감사합니다.")

# 뼈대 만들기
def calc_times_table(menu):    
    #1) 한 단
    if menu == '1':
        num = input("출력할 단을 입력하세요 : ")
            
        if num == 'q':
            return end()
            
        elif 0<int(num)<10:
            return calc_single_times_table(int(num))
                        
        elif int(num)<0 or int(num)>9:
            agreement = input(print("입력하신 숫자로는 구구단을 만들 수 없습니다. 정말 이대로 하시겠습니까? Y/N"))
                
            if agreement == 'Y':
                print("잘못된 입력으로 프로그램을 종료합니다.")
                    
            elif agreement == 'N':
                num = input("출력할 단을 입력하세요 : ")
                    
                if num == 'q':
                    return end()
                    
                elif 0<int(num)<10:
                    return calc_single_times_table(int(num))
                    
                else:
                    print("잘못된 입력입니다. 다시 입력해 주세요.")
                    return gugudan_main()
                
            elif agreement == 'q':
                return end()
            
            else:
                print("잘못된 입력입니다. 다시 입력해 주세요.")
                return gugudan_main()
            
        else:
            print("잘못된 입력입니다. 다시 입력해 주세요.")
            return gugudan_main()
        
    #2) 1 ~ n단
    elif menu == '2' :
        num = input("숫자를 입력하세요 : ")
        
        if num == 'q':
            return end()
    
        elif 0<int(num)<10 :
            return calc_ton_times_table(int(num))
            
        elif int(num)<0 or int(num)>9:
            agreement = input(print("입력하신 숫자로는 구구단을 만들 수 없습니다. 정말 이대로 하시겠습니까? Y/N"))
        
            if agreement == 'Y':
                print("잘못된 입력으로 프로그램을 종료합니다.")
                    
            elif agreement == 'N':
                num = input("숫자 입력하세요 : ")
                    
                if num == 'q':
                    return end()
                
                elif 0<int(num)<10:
                    return calc_ton_times_table(int(num))
                    
                else:
                    print("잘못된 입력입니다. 다시 입력해 주세요.")
                    return gugudan_main()
                    
            elif n == 'q' :
                return end()
            
            else:
                print("잘못된 입력입니다. 다시 입력해 주세요.")
                return gugudan_main()
        
        else :
            print("잘못된 입력입니다. 다시 입력해 주세요.")
            return gugudan_main()
    
    #3) n ~ m단
    elif menu == '3' : 
        num1 = input("숫자를 입력하세요 : ")
        num2 = input("숫자를 입력하세요 : ")
        
        if num1>num2:
            agreement = input(print("입력하신 숫자로는 구구단을 만들 수 없습니다. 정말 이대로 하시겠습니까? Y/N"))
        
            if agreement == 'Y':
                return calc_2to1_times_table(int(num2), int(num1))
                    
            elif agreement == 'N':
                num1 = int(input("숫자를 입력하세요 : "))
                num2 = int(input("숫자를 입력하세요 : "))
                
                return calc_1to2_times_table(int(num1), int(num2))
                        
            elif agreement == 'q':
                return end()
            
            else:
                print("잘못된 입력입니다. 다시 입력해 주세요.")
                return gugudan_main()
            
        elif num1<num2:
            return calc_1to2_times_table(int(num1), int(num2))
            
        elif n == 'q':
            return end()
             
        else :
            print("잘못된 입력입니다. 다시 입력해 주세요.")
            return gugudan_main()
                    
    elif menu == 'q' :
        return end()
    
    else :
        print("잘못된 입력입니다. 다시 입력해 주세요.")
        return gugudan_main()
        
    
# 프로그램 만들기
def gugudan_main() :
    print(
"""
구구단 출력기
------------------------------------------------------------------------------------
        1) n단 출력      2) n단까지 출력      3) n ~ m단 출력      q) 나가기 
------------------------------------------------------------------------------------
메뉴를 선택하세요 > 
""")
    
    menu = input()
    
    return calc_times_table(menu)
                      
gugudan_main()
```
