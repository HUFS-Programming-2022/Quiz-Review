----------------------------------------------------------------------
           1) n단 출력    2) n단까지 출력   3) n ~ m단 출력   q) 나가기
----------------------------------------------------------------------

```python
def print_gugudan(n, m):
    for dan in range(n, m+1):
        print('')
        print(f'====={dan}단=====')
        for i in range(1, 10):
            print(f'{dan} * {i} = {dan * i}')


def end_program():
    print('프로그램 종료')
    print('이용해주셔서 감사합니다.')

def n_dan():
    n = input('몇 단 출력?(1~9, 종료 = q)')
    if n == 'q':
        end_program()

    elif int(n) < 1 or int(n) > 9:
        n_dan()
    
    else :
        m = int(n)
        print_gugudan(int(n), int(m))
        gugudan()
        
def to_n():
    n = 1
    m = input('몇 단 까지?(1~9, 종료 = q)')
    
    if m == 'q':
        end_program()

    elif int(m) < 1 or int(m) > 9:
        to_n()
    else :
        print_gugudan(1, int(m))
        gugudan()

def n_to_m():
    n = input('몇 단 부터?(1~9, 종료 = q)')
    
    if n == 'q':
        end_program()
    elif int(n) < 1 or int(n) > 9:
        n_to_m()
    else :
        m = input('몇 단 까지?(1~9, 종료 = q)')
    
        if m == 'q':
            end_program() 
        elif int(m) <1 or int(m) > 9 or int(n) > int(m):
            n_to_m()
        else:
            print_gugudan(int(n), int(m))
            gugudan()


    
    
def gugudan():
        print('')
        print('''구구단 출력기
----------------------------------------------------------------------
           1) n단 출력    2) n단까지 출력   3) n ~ m단 출력   q) 나가기
----------------------------------------------------------------------
              ''')
        menu = input('메뉴를 선택하세요 > ')
        if menu == '1':
            n_dan()
        
        elif menu == '2':
            to_n()
        
        elif menu == '3':
            n_to_m()
        
        elif menu == 'q':
            end_program()
        else:
            gugudan()

gugudan()
```
