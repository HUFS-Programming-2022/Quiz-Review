```python
#gugudan
# 1. 구구단 계산기
# 1.1 메뉴 제시, 메뉴 번호 입력받기
print('"""')
print('구구단 출력기')
print('----------------------------------------------------------------------')
print('           1) n단 출력    2) n단까지 출력   3) n ~ m단 출력   q) 나가기')
print('----------------------------------------------------------------------')
print('"""')
# 1.2 메뉴 번호 입력받기, breaker 설정(메뉴 입력 후에도 q를 눌러 종료해야 할 때 사용됨)
breaker = True
while True :
    user_input = input("메뉴를 선택하세요>")
    if user_input=='q' :
        print("이용해주셔서 감사합니다.")
        break

# 1.3 입력된 메뉴 버튼에 따라 계산기 실행, 한 번 정상적으로 계산 마칠 시 다시 메뉴 란으로 돌아가서 q 입력되기 전까지 반복     
    elif user_input=='1' :
        while True :
            n = input("출력을 원하는 단 수를 입력하세요")
        
            
            if n=='q' : #조건 3
                print("이용해주셔서 감사합니다.")
                breaker = False
                break
                
            elif n.isdigit()==False : #예외사항 대처- 입력된 단 수가 숫자가 아닌 경우
                print("숫자를 입력해야 합니다. \n다시 시도하세요")
                continue
            else :
                num_n = int(n)
                if 0 > num_n or num_n > 9 : #조건 1
                    print("올바른 숫자가 아닙니다. \n1 ~ 9까지의 수를 입력하세요")
                    continue
                    
                else :
                    print(f'-------{num_n} 단------')
                    for i in range (1,10) :
                        print(f'{num_n}x{i}={num_n*i}')
                break
            
            
            
    elif user_input=='2' :
        while True :
            n = input("단 수를 입력하세요")

            if n == 'q' : #조건 3
                print("이용해주셔서 감사합니다.")
                breaker = False
                break
            elif n.isdigit()==False :
                print("숫자를 입력해야 합니다. \n다시 시도하세요")
                continue
            else :
                num_n = int(n)
                if 0 > num_n or num_n > 9 : #조건 1
                    print('올바른 숫자가 아닙니다.\n1 ~ 9까지의 수를 입력하세요')
                    continue
                
                else :
                    for i in range (1,num_n+1) :
                        print(f'-------{i} 단------')
                        for j in range (1,10) :
                            print(f'{i}x{j}={j*i}')
                        print()
                break
            
    elif user_input=='3' :
        while True :
            n = input("첫 단 수를 입력하세요")

            if n == 'q': #조건 3, 첫 단 수 먼저 조건 확인 후 마지막 단 입력받기
                print("이용해주셔서 감사합니다.")
                breaker = False
                break
            elif n != 'q' :
                if n.isdigit()== False :
                    print("숫자를 입력해야 합니다. \n다시 시도하세요.")
                    continue
                elif n.isdigit() :
                    m = input("마지막 단을 입력하세요")
                    if m == 'q' :
                        print("이용해주셔서 감사합니다.")
                        breaker = False
                        break
                    elif m.isdigit() == False :
                        print("숫자를 입력해야 합니다. \n다시 시도하세요.")
                        continue
                    elif m.isdigit() :
                            num_n = int(n)
                            num_m = int(m) #조건 1
                            if not 0 < num_n < 10 or not 0 < num_m < 10:
                                print('올바른 숫자가 아닙니다.\n1 ~ 9 사이의 수를 입력하세요')
                                continue
                            elif num_n > num_m : #조건2
                                print('첫 단 수는 마지막 단 수보다 클 수 없습니다. \n다시 시도하세요.')
                                continue
                            else :
                                for i in range (num_n,num_m+1) :
                                    print(f'-------{i} 단------')
                                    for j in range (1,10) :
                                        print(f'{i}x{j}={j*i}')
                                        print()
                            break
                                
#예외 대처- 최초 메뉴 입력값이 1,2,3,q 중 어느 것도 아닐 경우            
    elif user_input != '1' or '2' or '3' or 'q' :
        print("입력이 잘못되었습니다. 다시 시도하세요.")
        continue
        
    if breaker==False :
        break
```
