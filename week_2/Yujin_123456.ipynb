```python
#2-2
text4 = book.text4.tokens
text5 = book.text5.tokens
fdist = nltk.FreqDist(text4)
once=fdist.hapaxes()
len(set([word for word in text5 if word in once and text5.count(word) >= 2]))
```

```python
#3
import nltk
from nltk import book
text1 = book.text1
text2 = book.text2
text3 = book.text3
text4 = book.text4
text5 = book.text5
text6 = book.text6
text7 = book.text7
text8 = book.text8
text9 = book.text9
text=[text1, text2, text3, text4, text5, text6, text7, text8, text9]

text_diversity=[]
for moby in text:    
    text_diversity.append(len(set(moby)) / len(moby))
print(max(text_diversity))
```
