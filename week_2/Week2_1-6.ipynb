## 1. text5에서 4번 출현하는 단어들을 중복되지 않게 오름차순으로 출력하시오

```python
# 4번 출현하는 단어들을 찾기 위해 FreqDist 사용한다
fdist5 = FreqDist(text5)

# 리스트 컴프리헨션을 사용해 원하는 조건의 단어를 찾는다
# 조건: fdist를 사용해 4번 출현하는 단어
# 조건: 알파벳으로 이루어진 단어
# set()으로 중복을 없애준다
# sorted()로 오름차순 정렬을 해준다
# (대소문자를 구분하지 않았다)
sorted(set([word for word in text5 if fdist5[word] == 4 and word.isalpha()]))
```

## 2-2. text4에서는 한 번 출현하지만 text3에서는 여러 번 출현하는 단어는 몇 개인지 출력하시오

```python
# 단어 출현 개수를 찾기 위해 FreqDist를 사용한다
fdist4 = FreqDist(text4)
fdist3 = FreqDist(text3)

# 리스트 컴프리헨션을 사용해 원하는 조건의 단어를 찾는다
# 조건: 알파벳으로 이루어진 단어
# 조건: text4에서 한 번 출현하는 단어
# 조건: text3에서 여러 번 출현하는 단어
# len()를 사용해 모든 조건에 맞는 단어의 개수를 구한다
len([word for word in text4 if word.isalpha() and fdist4[word] == 1 and fdist3[word] > 1])
```

## 3. text1~9 중 가장 어휘가 풍부한 코퍼스는?

```python
# 모든 텍스트를 리스트에 담아준다
texts = [text1, text2, text3, text4, text5, text6, text7, text8, text9]

# 각 텍스트의 어휘 비율을 담을 text_ratio 빈 리스트를 생성한다
text_ratio = []

# for문을 사용해 각 텍스트의 비율을 계산한 뒤 text_ratio에 담아준다
for txt in texts:
    text_ratio.append(len(set(txt)) / len(txt))

# max()를 이용해 비율이 가장 높은 텍스트를 찾는다
# index()를 이용해 해당 텍스트가 몇 번째인지 찾는다
# 0부터 시작하므로 1을 더해준 뒤 해당 텍스트를 출력한다
text_number = text_ratio.index(max(text_ratio)) + 1
print(f'가장 어휘가 풍부한 코퍼스는 text{text_number} 입니다.')
```

## 5. text7에서 가장 긴 단어는 몇 자(character)인가?
