### 4. NLTK 임포트하고, text4에서 4자리 단어만 출력하세요.

```python
from nltk import book

address = book.text4
four_word = []

for w in address.tokens:
    if len(w) == 4 and w.isalpha():
        four_word.append(w)
        
sorted(set(four_word))
```
